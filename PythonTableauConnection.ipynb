# Q3) Divide ABS BR into different bins

```python
print ("Maximum value in Abs birth rate is of India in 2009 which is ")
print (df_5years["ABS_Birth_Rate"].max())
# print(df_5years[df_5years['ABS_Birth_Rate'] == df_5years['ABS_Birth_Rate'].max()])
# print ("Minimum value in Abs birth rate is of San Mariono in 2011 which is ") 
# print (df_5years["ABS_Birth_Rate"].min())
# print(df_5years[df_5years['ABS_Birth_Rate'] == df_5years['ABS_Birth_Rate'].min()])
print (df_5years["ABS_Birth_Rate"].describe())
import seaborn as sns
sns.boxplot(df_5years["ABS_Birth_Rate"])
```

```python
df_5years["BR.Bins"]=np.where(df_5years["ABS_Birth_Rate"]<700,"Below 700", "Above 700")
condition= [(df_5years["ABS_Birth_Rate"]<=10),
        (df_5years["ABS_Birth_Rate"]>10) & (df_5years["ABS_Birth_Rate"]<=50),
        (df_5years["ABS_Birth_Rate"]>50) & (df_5years["ABS_Birth_Rate"]<=129),
        (df_5years["ABS_Birth_Rate"]>129) & (df_5years["ABS_Birth_Rate"]<=578),
        (df_5years["ABS_Birth_Rate"]>578) & (df_5years["ABS_Birth_Rate"]<=3000),
        (df_5years["ABS_Birth_Rate"]>3000) & (df_5years["ABS_Birth_Rate"]<=7000),
        (df_5years["ABS_Birth_Rate"]>7000), df_5years["ABS_Birth_Rate"].isnull()]
values=["Very Low", "Low", "Below Avg", "Avg", "Above Avg", "High", "Very High", "Missing"]
df_5years["BR.Bins.New"]=np.select(condition,values)
print(df_5years)
df_5years.to_csv("/Users/sunitjadeja/Desktop/Python/World Indicators Extract_Full Data (3).csv")
```
